```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def euler(function, y0 = 0, x_interval = (0,1), stepsize = 0.1):
    """returns the solution of a differential equation with Euler method

    Keyword arguments:
    function -- the differential equation to solve dy/dx = f(x,y)
    y0 -- beginning value of y0 (default = 0)
    x_interval -- x interval to solve (default = (0,1))
    stepsize -- stepsize for Euler (default = 0.1)
    """
    xaxis = np.arange(x_interval[0],x_interval[1],stepsize)
    y=[y0]
          
    for x in xaxis[:-1]:
        y.append(y[-1]+function(x,y[-1])*stepsize)
        
    return(xaxis,y)

def f(x,y):
    return 2*y+x

plt.plot(*euler(f))
    
```
